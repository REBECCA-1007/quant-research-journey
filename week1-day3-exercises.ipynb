```python
# bài tập 1: temperature converter funtion 
# yêu cầu:
# viết 2 functions:
# celsius_to_fahrenheit(c) - convert C sang F
# fahrenheit_to_celsius(f) - convert F sang C
# formula:
# F = C x 9/5 + 32
# C = (F - 32) x 5/9 
# test với nhiều giá trị 

def celsius_to_fahrenheit (C):
    F = C * 9/5 + 32
    return F
def fahrenheit_to_celsius (F):
    C = (F - 32) * 5/9
    return C
print(celsius_to_fahrenheit(0))
print(celsius_to_fahrenheit(100))
print(fahrenheit_to_celsius(32))
print(fahrenheit_to_celsius(212))
```

```python
# bài tập 2: list statistics function
# yêu cầu
# viết function get_list_stats(numbers)
# return: min, max, sum, average 
# test với list khác nhau 

def get_list_stats(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    total = sum(numbers)
    avg = total /len (numbers)
    return min_val, max_val, total, avg
nums = [5,2,8,1,9,3]
min_val, max_val, total, avg = get_list_stats(nums)
print(f"Min: {min_val}, Max: {max_val}, Sum: {total}, Average: {avg}")
```

```python
# bài tập 3: todo list manager 
# yêu cầu: 
# tạo empty list todos
# viết 4 function:
# add_todo(task) - thêm task
# remove_todo(task) - xóa task
# show_todos() - hiện tất cả tasks
# mark_done(index) - đánh dấu task hoàn thành (thêm "✓")

todos = []
def add_todo(task):
    todos.append(task)
    print(f"✓ Đã thêm: {task}")
def remove_todo(task):
    todos.remove(task)
    print(f"✓ Đã xóa: {task}")
def show_todos():
    if len(todos) == 0:
        print("Danh sách trống")
    else: 
        print("\nDanh sách TODO:")
        for i, todo in enumerate(todos, start=1):
            print(f"{i}. {todo}")
def mark_done(index):
    todos[index] = "✓" + todos[index]
    print(f"✓ Đã hoàn thành: {todos[index]}")

add_todo("Học Python")
add_todo("Làm bài tập")
add_todo("Commit GitHub")
remove_todo("Làm bài tập")
show_todos()
mark_done(0)
show_todos()
```

```python
# bài tập 4: find primes in range 
# yêu cầu: 
# viết function is_prime(n) - check số nguyên tố
# viết function get_primes(start, end) - tìm tất cả số nguyên tố trong range 
# return list các số nguyên tố 

def is_prime(n):
    if n < 2: return False
    for i in range (2,n):
        if n % i == 0: 
            return False
    return True 
def get_primes(start, end):
    primes = [i for i in range(start, end) if is_prime(i)]
    return primes

print(is_prime(7))
print(is_prime(10))
print(get_primes(1,20))
```

```python
# bài tập 5: stock price analyzer
# yêu cầu:
# cho list giá cổ phiếu theo ngày 
# viết function analyze_stock(prices):
# highest price
# lowest price
# average price
# best day to buy (lowest)
# best day to sell (highest)
# profit if buy lowest, sell highest

def analyze_stock(prices):
    max_val = max(prices)
    min_val = min(prices)
    avg = sum(prices) / len(prices)
    buy_on_day = prices.index(min_val)
    sell_on_day = prices.index(max_val)
    max_profit = max_val - min_val
    return max_val, min_val, avg, buy_on_day, sell_on_day, max_profit

prices = [100,95,103,98,110,92,105]
result = analyze_stock(prices)

print(f"Highest: {result[0]}")
print(f"Lowest: {result[1]}")
print(f"Average: {result[2]}")
print(f"Buy on day: {result[3]}")
print(f"Sell on day: {result[4]}")
print(f"Max profit: {result[5]}")
```
