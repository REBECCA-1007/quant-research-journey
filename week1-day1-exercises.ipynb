```python
# Bài 1: Calculator 
# Yêu cầu: 
# Nhập 2 số 
# Tính tổng, hiệu, tích, thương 
# In kết quả
```

```python
num1 = float(input("Nhập số thứ nhất: "))
num2 = float(input("Nhập số thứ hai: "))

Tong = num1 + num2
Hieu = num1 - num2 
Tich = num1 * num2
Thuong = num1 / num2

print("Nhập số thứ nhất", num1)
print("Nhập số thứ 2:", num2)
print ("Tổng", Tong)
print ("Hiệuu", Hieu)
print ("Tích", Tich)
print ("Thương", Thuong)
```

```python
# Bài 2: BMI calculator
# Yêu cầu:
# BMI = weight/(height)^2
# In kết quả
```

```python
weight = float(input("Nhập số cân nặng: "))
height = float(input("Nhập số chiều cao: "))

BMI = weight / (height) ** 2

print ("Nhập số cân nặng(kg): ", weight)
print ("Nhập số chiều cao(m): ", height)
print ("BMI: ", BMI)
```

```python
# Bài 3: Temperature Converter
# Yêu cầu: Convert Celsius Fahrenheit
# F = C x 9/5 + 32
```

```python
C = float(input("Nhập số độ C: "))
F = C * 9/5 + 32
print ("Nhập số độ C: ", C)
print ("Nhiệt độ F sau chuyển đổi là: ", F)
```

```python
# Bài 5: Interest Calculator
# Yêu cầu: Tính lãi đơn giản 
# Interest = Principal x Rate x Time/100
```

```python
principal = float(input("Nhập số tiền gốc:"))
rate = float(input("Nhập lãi suất (%/năm)"))
time = float(input("Nhập số năm:"))
interest = principal * rate * time/100
print ("Số vốn gốc:", principal)
print ("Lãi suất:", rate)
print ("Số năm:", time)
print ("Số lãi cần trả là:", interest)
```
