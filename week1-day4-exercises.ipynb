```python
# bài tập 1: student grade manager
# yêu cầu:
# tạo dictionary lưu thông tin học sinh
# functions:
# add_student(name, grades) - thêm học sinh 
# get_average(name) - tính điểm trung bình
# get_top_student() - tìm học sinh điểm cao nhất
# save_to_file(filename) - lưu vào file

students = {}
def add_student(name, grades):
    students[name] = grades
def get_average(name):
    if name in students:
        return sum(students[name]) / len(students[name])
    return None
def get_top_student():
    if not students:
        return None
    top_student = None
    highest_avg = 0
    for name, grades in students.items():
        avg = sum(grades) / len(grades)
        if avg > highest_avg:
            highest_avg = avg 
            top_student = name
    return top_student
def save_to_file(filename):
    with open(filename, "w") as file:
        for name, grades in students.items():
            avg = sum(grades) / len(grades)
            file.write(f"{name}: {grades} (Avg: {avg:.2f})\n")

add_student("Rebecca", [95,87,92])
add_student("Florence", [88,91,85])
add_student("Alex", [78,82,80])

print(f"Rebecca's average: {get_average('Rebecca')}")
print(f"Top student: {get_top_student()}")
save_to_file("students.txt")
```

```python
# bài tập 5: stock porfolio tracker
# yêu cầu:
# track porfolio với nested dict
# structure: {symbol: {"shares": x, "buy_price": y}}
# functions:
# add_positon(symbol, shares, buy_price) - thêm cổ phiếu
# calculate_porfolio_value(current_prices) - tính giá trị
# calculate_profit_los(current_prices) - tính lời lỗ
# save_porfolio(filename) - lưu JSON
# load_porfolio(filename) - load JSON

import json
portfolio = {}
def add_position(symbol, shares, buy_price):
    portfolio[symbol] = {"shares": shares, "buy_price": buy_price}
def calculate_portfolio_value(current_prices):
    total_value = 0
    for symbol, data in portfolio.items():
        shares = data["shares"]
        current_price = current_prices.get(symbol, 0)
        value = shares * current_price
        total_value += value
    return total_value

def calculate_portfolio_loss(current_prices):
    total_pl = 0
    for symbol, data in portfolio.items():
        shares = data["shares"]
        buy_price = data["buy_price"]
        current_prices = current_prices.get(symbol,0)
        pl = (current_prices - buy_price) * shares
        total_pl += pl
    return total_pl

def save_portfolio(filename):
    with open(filename, "w") as file:
        json.dump(portfolio, file, indent=2)

def load_portfolio(filename):
    global portfolio
    with open(filename, "r") as file:
        data = json.load(file)
    print(data["symbol"])

add_position("BTC", 0.5, 40000)
add_position("ETH", 2, 2000)
current_prices = {"BTC": 45000, "ETH": 2500}
print(f"Portfolio value: ${calculate_portfolio_value(current_prices)}")
print(f"Profit/Loss: ${calculate_portfolio_loss(current_prices)}")
save_portfolio("portfolio.json")
```
