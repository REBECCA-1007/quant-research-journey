```python
# bài tập 1: student grade manager
# yêu cầu:
# tạo dictionary lưu thông tin học sinh
# functions:
# add_student(name, grades) - thêm học sinh 
# get_average(name) - tính điểm trung bình
# get_top_student() - tìm học sinh điểm cao nhất
# save_to_file(filename) - lưu vào file

students = {}
def add_student(name, grades):
    students[name] = grades
def get_average(name):
    if name in students:
        return sum(students[name]) / len(students[name])
    return None
def get_top_student():
    if not students:
        return None
    top_student = None
    highest_avg = 0
    for name, grades in students.items():
        avg = sum(grades) / len(grades)
        if avg > highest_avg:
            highest_avg = avg 
            top_student = name
    return top_student
def save_to_file(filename):
    with open(filename, "w") as file:
        for name, grades in students.items():
            avg = sum(grades) / len(grades)
            file.write(f"{name}: {grades} (Avg: {avg:.2f})\n")

add_student("Rebecca", [95,87,92])
add_student("Florence", [88,91,85])
add_student("Alex", [78,82,80])

print(f"Rebecca's average: {get_average('Rebecca')}")
print(f"Top student: {get_top_student()}")
save_to_file("students.txt")
```

```python
# bài tập 2: word frequency counter
# yêu cầu:
# đọc text file 
# đếm số lần xuất hiện của mỗi từ
# return dictionary {word:count}
# in top 5 từ xuất hiện nhiều nhất 

def count_words(filename):
    word_counts = {}
    try:
        with open(filename, "r", encoding="utf-8") as file:
            for line in file: # xử lý từng dòng
                line = line.strip().lower() # chuyển về lowercase
                for char in ".,!?;:\"'": # xóa dấu câu .,!?;:
                    line = line.replace(char,"")
                words = line.split() # split thành words
                for word in words: # đếm từng word
                    if word: # skip empty strings
                        word_counts[word] = word_counts.get(word,0) + 1 # dùng get()
        return word_counts
    except FileExistsError:
        print(f"File {filename} not found!")
        return {}

def get_top_words(word_counts, n=5):
    sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse = True)
    return sorted_words[:n]

test_content = """
Python is awesome. Python is powerful. 
I love Python. Python Python Python!
Learning Python is fun and Python makes coding easy.
"""
with open("test.txt", "w", encoding="utf-8") as f:
    f.write(test_content)
print("=== Word Frequency Counter ===\n")

word_counts = count_words("test.txt")
print(f"Total unique words: {len(word_counts)}")
print(f"Total words counted: {sum(word_counts.values())}\n")

top_5 = get_top_words(word_counts, 5)
print("Top 5 most frequent words:")
for word, count in top_5:
    print(f"{word}: {count}")
print("\n--- All words by frequency ---")
all_sorted = get_top_words(word_counts, len(word_counts))
for word, count in all_sorted:
    print(f"{word}: {count}")
```

```python
# bài tập 3: phonebook
# yêu cầu: 
# tạo phonebook với dictionary 
# functions: 
# add_contact(name, phone, email) - thêm contact
# search_contact(name) - tìm contact
# update_contact(name, phone=None, email=None) - update
# delete_contact(name) -  xóa
# list_all_contatc() - hiện tất cả

phonebook = {}
def add_contact(name, phone, email):
    phonebook[name] = {"phone": phone, "email": email}
def search_contact(name):
    if name in phonebook:
        contact = phonebook[name]
        print(f"Name: {name}")
        print(f"Phone: {contact['phone']}")
        print(f"Email: {contact['email']}")
        return contact
    else:
        print(f"Contact not found")
        return None

def update_contact(name, phone=None, email=None):
    if name not in phonebook:
        print("Not found")
        return
    if phone is not None:
        phonebook[name]["phone"] = phone
    if email is not None: 
        phonebook[name]["email"] = email
    print(f"Updated: {name}")

def delete_contact(name):
    if name in phonebook:
        del phonebook[name]
        print(f"Deleted: {name}")
    else:
        print("Not found")

def list_all_contact():
    if not phonebook:
        print("Phonebook is empty")
        return
    print("\n=== PHONEBOOK ===")
    for name, contact in phonebook.items():
        print(f"\nName: {name}")
        print(f" Phone: {contact['phone']}")
        print(f" Email: {contact['email']}")
    print("============\n")


add_contact("Rebecca", "0389696946","becca@gmail.com")
add_contact("Florence", "0373081837", "florence@gmail.com")
add_contact("Alex", "0335329160", "alex@gmail.com")
list_all_contact()
print(search_contact("Rebecca"))
update_contact("Rebecca", phone="0335329160")
delete_contact("Alex")
list_all_contact()
```

```python
# bài tập 4: config file manager
# yêu cầu: 
# đọc config từ file
# parse thành dictionary
# allow update và save back
# format: key=value mỗi dòng

def load_config(filename):
    config = {}
    try: 
        with open(filename, "r") as file:
            for line in file: 
                line = line.strip() # xóa whitespace/ newline 
                if not line or line.startswith("#"): # skip empty lines và comments
                    continue
                if "=" in line: # parse key=value
                    key, value = line.split("=",1) # split tại dấu = đầu tiên 
                    config[key.strip()] = value.strip()
        return config
    except FileNotFoundError:
        print(f"File {filename} not found!")
        return {}
    
def save_config(config, filename):
    with open(filename, "w") as file:
        for key, value in config.items():
            file.write(f"{key}={value}\n")
    print(f"✓ Config saved to {filename}")

def get_config_value(config, key, default=None):
    return config.get(key, default)

initial_config = {
    "host": "{localhost}",
    "port": "{8080}",
    "debug": "{True}"
}
save_config(initial_config, "config.txt")
print("=== Test Config Manager ===\n")
config = load_config("config.txt")
print("Loaded config:", config)
print(f"Host: {get_config_value(config, 'host')}")
print(f"Port: {get_config_value(config, 'port')}")
print(f"Debug: {get_config_value(config, 'debug')}")
print(f"Timeout: {get_config_value(config, 'timeout', '30')}")
print("\n--- Moidifying config ---")
config["port"] = "9000"
config["timeout"] = "60"
save_config(config,"config.txt")
print("\n--- Verify changes---")
config = load_config("config.txt")
print("Config after changes:", config)
```

```python
# bài tập 5: stock porfolio tracker
# yêu cầu:
# track porfolio với nested dict
# structure: {symbol: {"shares": x, "buy_price": y}}
# functions:
# add_positon(symbol, shares, buy_price) - thêm cổ phiếu
# calculate_porfolio_value(current_prices) - tính giá trị
# calculate_profit_los(current_prices) - tính lời lỗ
# save_porfolio(filename) - lưu JSON
# load_porfolio(filename) - load JSON

import json
portfolio = {}
def add_position(symbol, shares, buy_price):
    portfolio[symbol] = {"shares": shares, "buy_price": buy_price}
def calculate_portfolio_value(current_prices):
    total_value = 0
    for symbol, data in portfolio.items():
        shares = data["shares"]
        current_price = current_prices.get(symbol, 0)
        value = shares * current_price
        total_value += value
    return total_value

def calculate_profit_loss(current_prices):
    total_pl = 0
    for symbol, data in portfolio.items():
        shares = data["shares"]
        buy_price = data["buy_price"]
        current_price = current_prices.get(symbol,0)
        pl = (current_price - buy_price) * shares
        total_pl += pl
    return total_pl

def save_portfolio(filename):
    with open(filename, "w") as file:
        json.dump(portfolio, file, indent=2)

def load_portfolio(filename):
    global portfolio
    with open(filename, "r") as file:
        portfolio = json.load(file)
  

add_position("BTC", 0.5, 40000)
add_position("ETH", 2, 2000)
current_prices = {"BTC": 45000, "ETH": 2500}
print(f"Portfolio value: ${calculate_portfolio_value(current_prices)}")
print(f"Profit/Loss: ${calculate_profit_loss(current_prices)}")
save_portfolio("portfolio.json")
```
