```python
# Bài tập 1: Grade calculator
# yêu cầu: 
# nhập điểm (0-100)
# in ra grade (A,B,C,D,F)
# A: 90-100, B: 80-89, C: 70-79, D: 60-69, F <60

score = float(input("Nhập số điểm (0-100):"))
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"
print (f"Grade: {grade}")
```

```python
# Bài tập 2: Even Odd
# Yêu cầu:
# Nhập 1 số:
# Check số chẵn hay lẻ 
# In kết quả 
num = int(input("Nhập một số:"))
if num % 2 == 0: 
    print (f"{num} là số chẵn")
else: 
    print (f"{num} là số lẻ")
```

```python
# Bài tập 3: Leap Year Checker
# yêu cầu: 
# nhập năm
# check năm nhuận hay không
# rules: 
# chia hết cho 4  + không chia hết cho 100 -> nhuận
# chia hết cho 400 -> nhuận

year = int(input("Nhập năm:"))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 ==0):
    print (f"{year} là năm nhuận")
else:
    print (f"{year} là năm không nhuận")
    
```

```python
# bài tập 4: Multiplicatiopn Table
# yêu cầu: 
# nhập 1 số
# in bảng cửu chương của số đo (1-10)

n = int(input("Nhập 1 số:"))
for i in range (1,11): 
    result = n * i
    print (f"{n} x {i} = {result}")
```

```python
# bài tập 5: sum of numbers 
# yêu cầu 
# nhập 1 số N 
# tính tổng từ 1 đến N
# in kết quả 

n = int(input("Nhập N:"))
sum = (n*(n+1))/2
print (f"Tổng từ 1 đến {n} là:", sum)
```

```python
# bài tập 5: cách 2
n = int(input("Nhập N:"))
sum = 0
for i in range (1, n+1):
    sum += i 
print (f"Tổng từ 1 đến {n} là: {sum}")
```

```python
# bài tập 5: cách 3
n = int(input("Nhập N:"))
sum = 0
i = 1 
while i <= n:
    sum += i 
    i +=1
print (f"Tổng từ 1 đến {n} là {sum}")
```
